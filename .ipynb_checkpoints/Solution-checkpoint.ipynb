```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('DataScienceChallenge.csv',index_col=None)
data['y'] = (data['y'] > 0).astype(int)
data['x23'] = (data['x23']=='TRUE').astype(int)
data['x2'] = data['x2'].str[0].astype('int')
data.fillna(0)
```

```python
pie_plt=data.groupby(['x15']).sum()['y'].sort_values(ascending=False)
pie_plt.columns=['brand','count']
pie_plt.head(20)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#correlation matrix heatmap
import seaborn as sns
correlation=data.corr() 
fig,axes=plt.subplots(figsize=(20,20))
cax=sns.heatmap(correlation,vmin=-0.25, vmax=1,square=True,annot=True)
#axes.set_xticklabels(['x1', 'x2', 'x3', 'x4','y'])
#axes.set_yticklabels(['x1', 'x2', 'x3', 'x4', 'y'])
axes.set_title('Heatmap of Correlation Matrix of numerical data')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
#X = np.matrix(data.loc[:, ['x2','x9']])
X = np.matrix(data.loc[:, ['x2']])
y = np.array(data['y'])
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
LR = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X_train, y_train)
LR.predict(X_test)
round(LR.score(X_test,y_test), 4)
```

```python
SVM = svm.SVC(decision_function_shape="ovo").fit(X_train, y_train)
SVM.predict(X_test)
round(SVM.score(X_test, y_test), 4)
```

```python
RF = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0).fit(X_train, y_train)
RF.predict(X_test)
round(RF.score(X_test, y_test), 4)
```

```python
NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(150, 10), random_state=1).fit(X_train, y_train)
NN.predict(X_test)
round(NN.score(X_test, y_test), 4)
```
